```python
import pandas as pd
```

```python
path_github = 'https://raw.githubusercontent.com/alexandergribenchenko/Test_R5_DE/main/data/dataset.csv'
```

```python
df = pd.read_csv(path_github, dtype=object)
df
```

# Pregunta 03.

## 03.01. ¿Cuáles son las tiendas con compras de al menos 100 clientes diferentes?

```python
df_Q_03_01 = df.groupby(['codigo_tienda'], as_index=False).\
           agg(n_unique_clientes=('num_documento_cliente', lambda x: len(set(x)))).\
           sort_values(by=['n_unique_clientes'], ascending=[False]).reset_index(drop=True)
df_Q_03_01 = df_Q_03_01[df_Q_03_01.n_unique_clientes>=100]
df_Q_03_01
```

## 03.02. ¿Cuáles son los 5 barrios donde la mayor cantidad de clientes únicos realizan compras en tiendas tipo 'Tienda Regional'?

```python
df_Q_03_02 = df[df.tipo_tienda=='Tienda Regional']
df_Q_03_02 = df_Q_03_02.groupby(['id_barrio', 'nombre_barrio'], as_index=False).\
           agg(n_unique_clientes=('num_documento_cliente', lambda x: len(set(x)))).\
           sort_values(by=['n_unique_clientes'], ascending=[False]).reset_index(drop=True)
df_Q_03_02 = df_Q_03_02.head()
df_Q_03_02
```
